```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import json
```

```python
r = requests.get("http://services.swpc.noaa.gov/text/aurora-nowcast-map.txt")
```

```python
soup = BeautifulSoup(r.text, "lxml")
```

```python
text = soup.find('p').text
```

```python
lines = text.split('\n')
```

```python
i = 0
while True:
    if lines[i].strip()[0] != "#":
        break
    i += 1
```

```python
data_lines = lines[i:-1]
data = list(map(lambda x: x.split("   ")[1:], data_lines))
```

```python
len(data)
```

```python
len(data[0])
```

```python
len(data)
```

```python
lat_inc = 0.3515625
```

```python
points = []
```

```python
for i, group in enumerate(data):
    for j in group:
        points.append((-90 + lat_inc*i, 0, float(j))
```

```python
points[1024]
```

```python
long_inc = 0.32846715
```

```python
mult = 0
new_points = []
for i, p in enumerate(points):
    if i and i % 512 == 0:
        mult += 1
    new_points.append([p[0], -180 + long_inc*mult, p[2]])
```
